```python
# Task 3: K-Means Clustering on Sentiment Dataset

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Step 2: Load Dataset
file_path = "3) Sentiment dataset.csv"  # make sure the file is in the same folder as this notebook
data = pd.read_csv(r"C:\Users\Al Noor\Desktop\3) Sentiment dataset.csv")

print("Dataset Preview:")
print(data.head())
```

```python
print(data.columns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(data["Text"])  

print("\nTF-IDF Matrix Shape:", X.shape)
```

```python
# Step 4: Dimensionality Reduction for Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())
```

```python
# Step 5: Elbow Method to find Optimal Clusters
wcss = []  # within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker="o")
plt.title("Elbow Method - Optimal K")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Step 6: Apply KMeans with Optimal Clusters (e.g., K=3)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Step 7: Visualization in 2D
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y_kmeans, palette="viridis")
plt.title("K-Means Clustering of Sentiment Text (2D PCA Projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
